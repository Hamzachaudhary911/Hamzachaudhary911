# ** Find Average % Using simple, Apply and Lambda Method**

**(i) Using Mean**

```python
data['%'].mean()
```

**(ii)Find Average Using Apply Method**

```python
def percentage(x):

   return x/2
```

```python
data['Average_%']=data['%'].apply(percentage)
data
```

```python
def percentage(x):
  return x//2
data['Average_%']=data['%'].apply(percentage)
data.head()
```
